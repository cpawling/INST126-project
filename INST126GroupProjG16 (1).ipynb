```
#Testcode
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#The above code is for importing different visualization tools
living = pd.read_csv("movehubqualityoflife.csv")
cost = pd.read_csv("movehubcostofliving.csv")

whole_data = pd.merge(living, cost)

#This code is to merge two data sets into one dataset
#coverting GBP to USD for all of the cost sections
whole_data["Cappuccino"] = whole_data["Cappuccino"] *1.24
whole_data["Wine"] = whole_data["Wine"] *1.24
whole_data["Gasoline"] = whole_data["Gasoline"] *1.24
whole_data["Avg Rent"] = whole_data["Avg Rent"] *1.24
whole_data["Avg Disposable Income"] = whole_data["Avg Disposable Income"] *1.24

#This code utilizes the built-in pandas method to round all of the columns that went to 4 decimal places during the conversion phase back to 2 decimals
entire_data = whole_data.round(2)

```
